# 2_1 物体検出とは

## 物体検出の概要

物体検出とは画像中に認識できるオブジェクトに対して枠線を囲い（この枠線をBBox(バウンディングボックス)という）それに対し信頼度を0~1で表現する。  

<img src="https://mmm-ssss.com/wp-content/uploads/2018/11/ssd_keras_pics.jpg"></img>

## 物体検出タスクのインプットとアウトプット

インプットは画像  
アウトプットは次の３つになる


1.   画像中に物体が存在するかを示すバウンディングボックスの位置と大きさ情報
2.   各バウンディングボックスが何の物体であるのか示すラベル情報（物体クラス＋背景(何もうつさない)クラス）
3.   その検出に対する信頼度


バウンディングボックスの情報は長方形の形です。これを表す表現は大きく２種類


1.   (xmin(左端)，ymin（下底）,xmax（右端）,ymax（上端）)
2.   (cx(中心座標のx),cy(中心座標のy),w(幅),h(高さ))







## VOCデータセット

２０種の物体クラスをもつ画像データ

## SSD物体検出の流れ

SSDでは何を出力するのか？  
X　BBoxを出力させる  
O　複数の定型的な長方形であるデフォルトボックス(DBox)を作成する。これをどのように変換すればBBoxに近づくか表す情報(オフセット情報)を出力する。
  
  
具体的な計算方法  
DBoxの情報が(cx_d,cy_d,w_d,h_d)であった場合、オフセット情報が(∆cx,∆cy,∆w,∆h)の4変数となる。SSDにおいてBBoxの情報は
$$
cx=cx_d(1+0.1 \Delta cx)\\
cy=cy_d(1+0.1 \Delta cy)\\
w=w_d*exp(0.2\Delta w)\\
h=h_d*exp(0.2\Delta h)
$$
として計算する。

## 画像データ、アノテーションファイルへのファイルパスのリストの作成

```python
# 学習、検証の画像データとアノテーションデータへのファイルパスリストを作成する
def make_datapath_list(rootpath):
  """
  データへのパスを格納したリストを作成する
  Parameters
  ----------
  roootpath:str
    データフォルダへのパス
  
  Returns
  ----------
  ret:train_img_list,train_anno_list,val_img_list,val_anno_list
  """
  # 画像ファイルとアノテーションファイルへのパスのテンプレート(%sを変えるだけであらゆる画像を取得可能)を作成する
  imgpath_templete=osp.join(rootpath,'JPEGImages','%s.jpg')
  annopath_templete=osp.join(rootpath,'Annotations','%s.xml')
  #訓練と検証、それぞれのファイルID（ファイル名）を取得
  train_id_names=osp.join(rootpath+'ImageSets/Main/train.txt')
  val_id_names=osp.join(rootpath+'ImageSets/Main/val.txt')
  #訓練データの画像ファイルとアノテーションファイルへのパスリストを作成する
  train_img_list=list()
  train_anno_list=list()
  for line in open(train_id_names):
    file_id=line.strip()#両端末尾を削除　ex)¥nabc¥t　-> abc
    img_path=(imgpath_templete%file_id)#画像へのパス.
    #imgpath_templete中の「%s」にfile_idを代入する
    anno_path=(annopath_templete%file_id)#アノテーションへのパス
    train_img_list.append(img_path)#リストに追加
    train_anno_list.append(anno_path)#リストに追加
  #検証データの画像ファイルとアノテーションファイルへのパスリスト作成
  val_img_list=list()
  val_anno_list=list()
  for line in open(val_id_names):
    file_id=line.strip()
    img_path=(imgpath_templete%file_id)
    anno_path=(annopath_templete%file_id)
    val_img_list.append(img_path)#リストに追加
    val_anno_list.append(anno_path)#リストに追加
  return train_img_list,train_anno_list,val_img_list,val_anno_list
```

## XML形式のアノテーションデータをリストに変換する

```python
#xml形式のアノテーションをPythonのリスト形式に変換する
class Anno_xml2list(object):
  """
  一枚の画像に対する「xml形式のアノテーションデータ」を、画像サイズで規格化してから
  リスト形式に変換する
  Attributes
  ==========
  classes:リスト
  VOCのクラス名を格納したリスト
  """
  def __init__(self,classes):
    self.classes=classes
  def __call__(self,xml_path,width,height):
    """
    一枚の画像に対する「xml形式のアノテーションデータ」を、画像サイズで正規化
    Parameters
    ----------
    xml_path:str
    xmlファイルへのパス
    width:int
    対象画像の幅
    height:int
    対象画像の高さ
    Returns
    ---------
    ret:[[xmin,ymin,xmax,ymax,label_ind],...]
    物体のアノテーションデータを格納したリスト。画像内の存在する物体数分だけの要素を持つ
    """
    #画像内の全ての物体のアノテーションをこのリストに格納する
    ret=[]
    #xmlファイルを読み込む
    xml=ET.parse(xml_path).getroot()
    """
    ET.parse():xmlデータの読み込み
    .getroot()：最上階層の要素を取得
    ex)<data>
          <a>abc</a>
          <b>123</b>
        </data>
      の場合<data>...</data>が得られる
    """
    #root.iter(str):str の要素を持つ自分より下の全ての階層（子または孫階層）から検索してitersion化する
    #似たような関数にfindall(str)があるがこれは直接の子階層のみ検索を行う
    for obj in xml.iter('object'):
      #アノテーションで検知がDifficultになっているものは除外する
      difficult=int(obj.find('difficult').text)
      if difficult ==1:
        continue
      #一つの物体に対するアノテーションを格納するリスト
      bndbox=[]

      name=obj.find('name').text.lower().strip()#物体名
      bbox=obj.find('bndbox')#バウンディングボックスの情報
      #アノテーションのxmin,ymin,xmax,ymaxを取得して、0~1に正規化する
      pts=['xmin','ymin','xmax','ymax']
      for pt in pts:
        #VOCは原点を（１、１）と置いている為、（0,0）になるように補正する
        cur_pixel=int(bbox.find(pt).text)-1
        #幅と高さで正規化
        if (pt=='xmin') or (pt=='xmax'):
          cur_pixel/=width
        elif (pt=='ymin') or (pt=='ymax'):
          cur_pixel/=height
        bndbox.append(cur_pixel)
      #オブジェクトの名前にあうclassesのindexを取得する
      label_idx=self.classes.index(name)
      bndbox.append(label_idx)
      #retにbndboxをappendする。[[1,2,3]]+[[2,2,2]]=[[1,2,3],[2,2,2]]となる
      ret+=[bndbox]
    return np.array(ret)
```

```python
class DataTransform():
  """
  画像アノテーションの前処理クラス。訓練と推論で異なる動作をする。
  画像のサイズを300X300にする。
  学習時はデータオーギュメンテーションを行う
  Attribute
  --------------
  input_size  : int
     リサイズ先の画像の大きさ
  color_mean  : (B,G,R)
      各色チャンネルの平均値
  """
  def __init__(self,input_size,color_mean):
    #Compose:処理の配列を入力。但し、入力配列内の処理は引数が（img, boxes=None, labels=None）で
    #返り値も（img, boxes, labels）であるものとする。そして__call__も（img, boxes=None, labels=None）
    #を引数に取り、配列内の処理を順次__call__を実行する.
    """
    ConvertFromInts(),    #ConvertFromInt:boxesとlabelsはそのままでimgのみfloatに変換する
    #ToAbsoluteCoords:いま現在でboxesは0~1に正規化されている.
    #入力された画像サイズに合うように幅と高さを掛ける.imgやlabelには変更は加えない
    ToAbsoluteCoords(),
    #ソースコード確認するべき
    PhotometricDistort(),
    #ランダムで拡大、サイズ超過分は引数で指定した値にて補完
    Expand(color_mean),
    #ある制限に基づいてランダムに画像をクリッピング
    RandomSampleCrop(),
    #画像をランダムに反転させる
    RandomMirror(),
    #アノテーションを1-0に正規化(画像サイズで割ってやる)
    ToPercentCoords(),
    #画像サイズをinput x inputのサイズに変更する
    Resize(input_size),
    #BGRの色の平均値を引き算する
    SubtractMeans(color_mean)
    """
    self.data_transforms={"train":Compose([
                                           ConvertFromInts(),  # intをfloat32に変換
                                           ToAbsoluteCoords(),  # アノテーションデータの規格化を戻す
                                           PhotometricDistort(),  # 画像の色調などをランダムに変化
                                           Expand(color_mean),  # 画像のキャンバスを広げる
                                           RandomSampleCrop(),  # 画像内の部分をランダムに抜き出す
                                           RandomMirror(),  # 画像を反転させる
                                           ToPercentCoords(),  # アノテーションデータを0-1に規格化
                                           Resize(input_size),  # 画像サイズをinput_size×input_sizeに変形
                                           SubtractMeans(color_mean)  # BGRの色の平均値を引き算
                                           ]),
                          "val":Compose([ConvertFromInts(),
                                         Resize(input_size),
                                         SubtractMeans(color_mean)])
                          }
  def __call__(self,img,phase,boxes,labels):
    """
    Prameters
    ==============
    img:np.array
    入力画像

    phase:str
    "tarin" or "val"
    
    boxes:array
    [[一つ目Box左上w,一つ目Box左上h,一つ目Box右下w,一つ目Box右下h],
    [2つ目のBoxの上記の配列],
    [3つ目],
    [４つ目]]

    labels:str
    何が発見されたか

    """
    return self.data_transforms[phase](img,boxes,labels)
```

## Datasetを作成

```python
#VOC2012のDarasetを作成する
class VOCDataset(data.Dataset):
  """
  VOC2021のDatasetを作成するクラス

  Attribute
  ==================
  img_list:リスト
  画像パスを格納したリスト
  anno_list:リスト
  アノテーションへのパスを格納したパス
  phase:"train"or"test"
  学習とテストを見分ける
  transform:object
  前処理クラスのインスタンス
  transform_anno:object
  xml形式のアノテーションをリストに変換するインスタンス
  """
  def __init__(self, img_list, anno_list, phase, transform, transform_anno):
    self.img_list = img_list
    self.anno_list = anno_list
    self.phase = phase  # train もしくは valを指定
    self.transform = transform  # 画像の変形
    self.transform_anno = transform_anno  # アノテーションデータをxmlからリストへ

  def __len__(self):
    '''画像の枚数を返す'''
    return len(self.img_list)

  def __getitem__(self, index):
    '''
    前処理をした画像のテンソル形式のデータとアノテーションを取得
    '''
    im, gt, h, w = self.pull_item(index)
    return im, gt

  def pull_item(self, index):
    '''前処理をした画像のテンソル形式のデータ、アノテーション、画像の高さ、幅を取得する'''
    #1.画像の読み込み
    image_file_path=self.img_list[index]
    img=cv2.imread(image_file_path)
    height,width,channels=img.shape

    #2.xml形式のアノテーション情報のリスト化
    anno_file_path=self.anno_list[index]
    anno_list=self.transform_anno(anno_file_path,width,height)
    
    #3.前処理を実施
    img,boxes,labels=self.transform(img,self.phase,anno_list[:,:4],anno_list[:,4])

    #色チャネルの順番をテンソルで扱えるように変換する
    #BGR→RGB
    #(height,width,channel)→(channel,height,width)
    """
    BGR→RGB:numpyからtensorに変換する際のfrom_numpy(img[:,:,(2,1,0)])で行っている
    (height,width,channel)→(channel,height,width)：tensor型の次元変換の関数である.permute()で行っている
    """
    img=torch.from_numpy(img[:,:,(2,1,0)]).permute(2,0,1)
    
    #BBoxとラベルをセットしたnp.arraryを作成する。
    """
    np.hstack((a,b)):引数内の配列を横方向に連結する
    np.expand_dims(a,index):配列に次元数を追加する
    ex)
      a=[[2,1,3],[2,1,3]]
      a.shape -> (2,3)
      np.expand_dim(a,0).shape ->(1,2,3)
      np.expand_dim(a,1).shape ->(2,1,3)
      np.expand_dim(a,2).shape ->(2,3,1)
    """
    gt=np.hstack((boxes,np.expand_dims(labels,axis=1)))
    return img,gt,height,width
```

# 2_3 DataLoaderの実装

## DataLoaderを作成する

普通のDataLoaderは出力が決まったサイズでしかとらないが今回の場合アノテーションは画像ごとに異なるサイズを出力する。<br>
このような場合,DataLoaderクラスのデータ取り出し用の関数である「collate_fn」を改良する必要がある。

```python
def od_collate_fn(batch):
    """
    Datasetから取り出すアノテーションデータのサイズが画像ごとに異なります。
    画像内の物体数が2個であれば(2, 5)というサイズですが、3個であれば（3, 5）など変化します。
    この変化に対応したDataLoaderを作成するために、
    カスタイマイズした、collate_fnを作成します。
    collate_fnは、PyTorchでリストからmini-batchを作成する関数です。
    ミニバッチ分の画像が並んでいるリスト変数batchに、
    ミニバッチ番号を指定する次元を先頭に1つ追加して、リストの形を変形します。
    """
    targets = []
    imgs = []
    for sample in batch:
        imgs.append(sample[0])  # sample[0] は画像imgです
        targets.append(torch.FloatTensor(sample[1]))  # sample[1] はアノテーションgtです

    # imgsはミニバッチサイズのリストになっています
    # リストの要素はtorch.Size([3, 300, 300])です。
    # このリストをtorch.Size([batch_num, 3, 300, 300])のテンソルに変換します
    imgs = torch.stack(imgs, dim=0)

    # targetsはアノテーションデータの正解であるgtのリストです。
    # リストのサイズはミニバッチサイズです。
    # リストtargetsの要素は [n, 5] となっています。
    # nは画像ごとに異なり、画像内にある物体の数となります。
    # 5は [xmin, ymin, xmax, ymax, class_index] です

    return imgs, targets
```

# 2_4 ネットワークモデルの実装

SSDとは最終的なBBox（バウンディングボックス）を得る為に、その元となる様々な大きさのDBoxを用意する、これを学習を通してBBoxに近づけていく事で学習を達成させる。どのようにこれを実装するか見ていく！！

## SSDネットワークモデルの概要

合計４つのConvを基本としたモデルを作成する<br>
- vgg(Chap1で使用したVGG-16を使用する)
- exstras(vggの結果をさらに畳み込む)
- loc(vgg,exstrasから得られた特徴マップからDBoxの位置を取得する)
- conf(vgg,exstrasから得られた特徴マップからDBoxの信頼性を取得する)

<h2>vgg</h2>  

入力画像　→        VGG      → sorce2(19x19)   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;L2Norm<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;↓<br>

```python
#34層にわたるvggモジュールを作成する
def make_vgg():
  layers=[]
  in_channels=3
  #vggモジュールで使用する畳み込み層やマックスプーリングのチャンネル数
  """
  M:デフォルト（floorモード{切り捨て}）のmaxpooling
  MC:Ceilモード（小数点切り上げ）のmaxpooling
  """
  cfg=[64,64,"M",128,128,"M",256,256,256,"MC",512,512,512,"M",512,512,512]
  for v in cfg:
    if v=="M":
      layers+=[nn.MaxPool2d(kernel_size=2,stride=2)]
    elif v=="MC":
      layers+=[nn.MaxPool2d(kernel_size=2,stride=2,ceil_mode=True)]
    else:
      con2d=nn.Conv2d(in_channels,v,kernel_size=3,padding=1)
      layers+=[con2d,nn.ReLU(inplace=True)]
      in_channels=v 
  pool5=nn.MaxPool2d(kernel_size=3,stride=1,padding=1)
  conv6=nn.Conv2d(512,1024,kernel_size=3,padding=6,dilation=6)
  conv7=nn.Conv2d(1024,1024,kernel_size=1)
  layers+=[pool5,conv6,nn.ReLU(inplace=True),conv7,nn.ReLU(inplace=True)]
  return nn.ModuleList(layers)
```

```python
def make_exstras():
  layers=[]
  in_channels=1024
  cfg=[256,512,128,256,128,256,128,256]
  layers+=[nn.Conv2d(in_channels,cfg[0],kernel_size=(1))]
  layers+=[nn.Conv2d(cfg[0],cfg[1],kernel_size=(3),stride=2,padding=1)]
  layers+=[nn.Conv2d(cfg[1],cfg[2],kernel_size=(1))]
  layers+=[nn.Conv2d(cfg[2],cfg[3],kernel_size=(3),stride=2,padding=1)]
  layers+=[nn.Conv2d(cfg[3],cfg[4],kernel_size=(1))]
  layers+=[nn.Conv2d(cfg[4],cfg[5],kernel_size=(3))]
  layers+=[nn.Conv2d(cfg[5],cfg[6],kernel_size=(1))]
  layers+=[nn.Conv2d(cfg[6],cfg[7],kernel_size=(3))]
  return nn.ModuleList(layers)
```

## モジュールlocとconfを実装する

```python
def make_loc_conf(num_class=21,bbox_aspect_num=[4,6,6,6,4,4]):
  """
  sorce1~6をそれぞれの畳み込み層によって処理するモデルを作成する関数
  
  Param
  ==================================
  num_class:int
  本データで検出するクラス21種
  bbox_aspect_num:array
  アスペクトを変更しながら製作するbbox

  Return
  ==================================
  model:nn.ModuleList
  作成されたモジュール
  """
  loc_layers=[]
  conf_layers=[]
  
  #VGGの２２層目、conf4_3(source1)に対する畳み込み層
  loc_layers+=[nn.Conv2d(512,bbox_aspect_num[0]*4,kernel_size=3,padding=1)]
  conf_layers+=[nn.Conv2d(512,bbox_aspect_num[0]*num_class,kernel_size=3,padding=1)]
  
  #VGGの最終層(source2)に対する畳み込み層
  loc_layers+=[nn.Conv2d(1024,bbox_aspect_num[1]*4,kernel_size=3,padding=1)]
  conf_layers+=[nn.Conv2d(1024,bbox_aspect_num[1]*num_class,kernel_size=3,padding=1)]

  #exstraの(source3)に対する畳み込み層
  loc_layers+=[nn.Conv2d(512,bbox_aspect_num[2]*4,kernel_size=3,padding=1)]
  conf_layers+=[nn.Conv2d(512,bbox_aspect_num[2]*num_class,kernel_size=3,padding=1)]

  #exstraの(source4)に対する畳み込み層
  loc_layers+=[nn.Conv2d(256,bbox_aspect_num[3]*4,kernel_size=3,padding=1)]
  conf_layers+=[nn.Conv2d(256,bbox_aspect_num[3]*num_class,kernel_size=3,padding=1)]

  #exstraの(source5)に対する畳み込み層
  loc_layers+=[nn.Conv2d(256,bbox_aspect_num[4]*4,kernel_size=3,padding=1)]
  conf_layers+=[nn.Conv2d(256,bbox_aspect_num[4]*num_class,kernel_size=3,padding=1)]

  #exstraの(source6)に対する畳み込み層
  loc_layers+=[nn.Conv2d(256,bbox_aspect_num[5]*4,kernel_size=3,padding=1)]
  conf_layers+=[nn.Conv2d(256,bbox_aspect_num[5]*num_class,kernel_size=3,padding=1)]

  return nn.ModuleList(loc_layers),nn.ModuleList(conf_layers)
  
```

## L2Norm層の実装

画像サイズ(n,a,b)『n:channel,(a,b):画像サイズ』  
a*bのセルに対して  
$$
{X_{i,n}}^{'}=\frac{X_{i,n}}{\sqrt{\sum_n X_{i,n}}}\quad(n:チャンネル,i:セル)
$$
チャンネルを通してL2Norm正規化する。これによってチャンネル間で著しく値が異なるようなセルができなくなる。

```python
#convC4_3からの出力をscale=20のL2Normで正規化する層
class L2Norm(nn.Module):

  def __init__(self,input_channels=512,scale=20):
    """
    L2Normを実装する
    Param
    ======================
    input_channel:int
    入力のチャンネル数
    scale:int
    重みの初期化に利用する
    """
    super().__init__()
    #モデルのパラメタとして扱えるオブジェクト
    self.weight=nn.Parameter(torch.Tensor(input_channels))
    self.scale=scale
    #weightをscaleの値にする関数
    self.reset_parameters()
    self.eps=1e-10
  def reset_parameters(self):
    """
    self.weightの値をself.scaleの値にする
    """
    #nn.init.constant_(weight,scale)はweightをscaleの値に置換する（初期化する関数）
    init.constant_(self.weight,self.scale)

  def forward(self,x):
    """
    38x38の特徴量に対して、512チャンネルにわたって2乗和のルートを求め、それを使用して各特徴量を正規化してから係数を掛け算する
    Param
    ==================
    x:torch.Tensor
    入力（層の結果）
    """
    #pow(exp):expの数値分乗算
    #sum(dim,keepdim):dim次元目の和を取り、keepdimをTrueにする事で次元数は削減されない。
    ##最終的なサイズは[batch_num,1,38,38]になる
    norm=x.pow(2).sum(dim=1,keepdim=True).sqrt()+self.eps
    x=torch.div(x,norm)
    #unsqeezeで次元を拡張している Ex)サイズ[512]にunseqeese(0)をすると[1,512]
    ##これによってweightの大きさはxと同じになる
    weights=self.weight.unsqueeze(0).unsqueeze(2).unsqueeze(3).expand_as(x)
    out=weights*x
    return out
```

## デフォルトボックスを実装


```python
#デフォルトボックスを出力する
class DBox(object):
  def __init__(self,cfg):
    """
    デフォルトボックスを作成するクラス
    Param
    =============================
    cfg:dict
    各種必要なデータが含まれている
    """
    super().__init__()#初期設定
    self.image_size=cfg["input_size"]#画像サイズ（1辺）300
    self.feature_maps=cfg["feature_maps"]#各ソースの特徴マッピングサイズ。Ex)[38,19,...]
    #このように特徴マップの一辺あたりのピクセル数を格納する。

    self.num_priors=len(cfg["feature_maps"])#ソース数(=6)
    self.steps=cfg["steps"]#DBoxのピクセルサイズ。Ex)[8,16,...]
    self.min_sizes=cfg["min_sizes"]#小さいDBoxの大きさを決める。Ex）[30, 60, …] 小さい正方形のDBoxのピクセルサイズ
    self.max_sizes=cfg["max_sizes"]#大きいDBoxの大きさを決める。Ex) [60, 111, …] 大きい正方形のDBoxのピクセルサイズ
    self.aspect_ratios=cfg["aspect_ratios"]#長方形のアスペクト比

  def make_dbox_list(self):
    """
    Dboxを作成する
    """
    mean=[]
    for k,f in enumerate(self.feature_maps):
      #fまでの数の２ペアの組み合わせを作る
      ##(i,j)は特徴マップの座標を現す.これを利用して、DBoxの中心座標を作成する。
      for i,j in product(range(f),repeat=2):
        #特徴量の画像サイズ
        ##元画像の大きさにDBoxの大きさを割ることでその大きさのDBoxを得る為の特徴マップの大きさを定義
        ##例えば、一辺8pixelのDBoxは画像で分けると37.5分割できる、少なくとも38x38に圧縮された
        ##特徴マップを使用することで作成することができる。
        ##300 / 'steps': [8, 16, 32, 64, 100, 300]＝[37.5 , 18.75 , 9.375 , 4.6875 , 3 , 1]
        f_k=self.image_size/self.steps[k]
        #DBoxの中心座標を0~1に正規化する
        cx=(j+0.5)/f_k
        cy=(i+0.5)/f_k
        #アスペクト比１(正方形)の小さいDBox
        #サイズはmin_sizesに格納,最小サイズは0~1に正規化する
        s_k=self.min_sizes[k]/self.image_size
        mean+=[cx,cy,s_k,s_k]
        #アス比１の最も大きいDBox
        #最も大きいサイズはmax_sizesに保存されている
        ##この計算は最小の辺と最大の辺での面積を√することで平均化、画像サイズで割る
        s_k_prime=sqrt(s_k*(self.max_sizes[k]/self.image_size))
        #print("√(",s_k,"*(",self.max_sizes[k],"/",self.image_size,"))=",s_k_prime)
        mean+=[cx,cy,s_k_prime,s_k_prime]
        #その他、アス比のDBox
        for ar in self.aspect_ratios[k]:
          #何でsqrt()なのか
          ##縦横にこれをかけたり割ったりしているので
          mean+=[cx,cy,s_k*sqrt(ar),s_k/sqrt(ar)]
          mean+=[cx,cy,s_k/sqrt(ar),s_k*sqrt(ar)]
    #DBoxをテンソルに変換torch.size([8732,4])
    output=torch.Tensor(mean).view(-1,4)
    #DBoxが画像の外に出て行かないように大きさを0~1にする
    output.clamp_(max=1,min=0)    
    return output
```

## 関数decodeを実装する

SSDの推論時のクラスDetectで使用する関数。この関数はDBox=(cx_d,cy_d,w_d,h_d)の座標とSSDモデルから求めたオフセット情報(∆cx,∆cy,∆w,∆h){DBoxとBBoxとの差分を表す}から、BBoxの座標情報を作成する関数
$$
cx=cx_d(1+0.1 \Delta cx)\\
cy=cy_d(1+0.1 \Delta cy)\\
w=w_d*exp(0.2\Delta w)\\
h=h_d*exp(0.2\Delta h)
$$
を計算していく。そんで表示形式は(cx,cy,w,h)→(xmin,ymin,xmax,ymax)形式に変換する

```python
#オフセット情報を使い、DBoxを使用してBBoxの座標にする
def decode(loc,dbox_list):
  """
  オフセット情報を使用して、DBoxを使用してBBoxの座標にする

  Param
  ===========================
  loc:array shape=[8732,4]
  SSDモデルで推論するオフセット情報
  dbox_list: array shape=[8732,4]
  DBoxの情報

  Returns
  ===========================
  boxes:[x,im,ymin,xmax,ymax]
  BBoxの情報
  """
  #DBoxは[cx,cy,w,h]で格納されている
  #locも[∆cx,∆cy,∆w,∆h]で格納されている
  boxes=torch.cat((
      dbox_list[:,:2]+loc[:,:2]*2*dbox_list[:,:2],
      dbox_list[:,2:]*torch.exp(loc[:,2:]*0.2)
  ),dim=1)
  #boxesのサイズはtorch.Size([8732,4])となります

  #BBoxの座標情報を[cx,cy,w,h]→[xmin,ymin,xmax,ymax]
  boxes[:,:2]-=boxes[:,2:]/2    #xmin,ymin
  boxes[:,2:]+=boxes[:,:2]    #xmax,ymax

  return boxes
```

一番確信度の高いBBox以外は邪魔なので削除する

```python
#Non_Maximum Suppressionを行う
def nm_suppression(boxes,scores,overlap=0.45,top_k=200):
    """
    Non_Maximum Suppressionを行う
    boxesのうち被りすぎ（overlap以上）のBBoxを削除する
    
    Param
    ===============================================
    boxes:[確信度の閾値(0.01)を超えたBBoxの数,4]
    BBoxの情報。
    scores:[確信度の閾値(0,01)を超えたBBoxの数]
    confの情報

    Returns
    ================================================
    keep:list
    confの降順にnmsを通過したindexが格納
    count:int
    nmsを通過したBBoxの数
    """
    #returnのひな型を作成
    count=0
    #scoreの同タイプ（Tensor）でサイズscores.size(0){確信度の閾値(0,01)を超えたBBoxの数}の0ベクトルTensorを作成
    keep=scores.new(scores.size(0)).zero_().long()
    #各BBoxの面積を計算する
    x1=boxes[:,0]
    y1=boxes[:,1]
    x2=boxes[:,2]
    y2=boxes[:,3]
    area=torch.mul(x2-x1,y2-y1)

    #BBoxの被り度合い（IOC）を計算に使用するひな形作成
    temp_x1=boxes.new()
    temp_y1=boxes.new()
    temp_x2=boxes.new()
    temp_y2=boxes.new()
    temp_w=boxes.new()
    temp_h=boxes.new()
    
    #scoreを昇順に並び替える.vには並び替えた後の配列、idxは並び替える前のインデックスを格納
    v,idx=scores.sort(0)

    #上位top-k個(200個)のBBoxのインデックスを取り出す（200個存在しない場合もある）
    idx=idex[-top_k:]

    #idxの要素数が0出ない限るループする(numelはtensorの要素数を取得する)
    while idx.numel() > 0:
        i = idx[-1]  # 現在のconf最大のindexをiに

        # keepの現在の最後にconf最大のindexを格納する
        # このindexのBBoxと被りが大きいBBoxをこれから消去する
        keep[count] = i
        count += 1

        # 最後のBBoxになった場合は、ループを抜ける
        if idx.size(0) == 1:
            break

        # 現在のconf最大のindexをkeepに格納したので、idxをひとつ減らす
        idx = idx[:-1]

        # -------------------
        # これからkeepに格納したBBoxと被りの大きいBBoxを抽出して除去する
        # -------------------
        # ひとつ減らしたidxまでのBBoxを、outに指定した変数として作成する
        ##tensor(第一引数)中の次元（第二引数）からインデックス（第三引数）を取り出す.取り出された結果はoutに導入する.
        torch.index_select(x1, 0, idx, out=tmp_x1)
        torch.index_select(y1, 0, idx, out=tmp_y1)
        torch.index_select(x2, 0, idx, out=tmp_x2)
        torch.index_select(y2, 0, idx, out=tmp_y2)

        
        ##最小値をx1[i]（最大confボックスの左端）に制限
        tmp_x1 = torch.clamp(tmp_x1, min=x1[i])
        ##最小値をx1[i]（最大confボックスの上端）に制限
        tmp_y1 = torch.clamp(tmp_y1, min=y1[i])
        ##最大値をx2[i]（最大confボックスの右端）に制限
        tmp_x2 = torch.clamp(tmp_x2, max=x2[i])
        ##最大値をｙ2[i]（最大confボックスの下端）に制限
        tmp_y2 = torch.clamp(tmp_y2, max=y2[i])

        # wとhのテンソルサイズをindexを1つ減らしたものにする
        ##引数に設定したtensorと同じサイズにtensorサイズを変更
        tmp_w.resize_as_(tmp_x2)
        tmp_h.resize_as_(tmp_y2)

        # clampした状態でのBBoxの幅と高さを求める
        tmp_w = tmp_x2 - tmp_x1
        tmp_h = tmp_y2 - tmp_y1

        # 幅や高さが負になっているものは0にする
        tmp_w = torch.clamp(tmp_w, min=0.0)
        tmp_h = torch.clamp(tmp_h, min=0.0)

        # clampされた状態での面積を求める
        inter = tmp_w*tmp_h

        # IoU = intersect部分 / (area(a) + area(b) - intersect部分)の計算
        rem_areas = torch.index_select(area, 0, idx)  # conf最大ボックス以外の各BBoxの元の面積
        union = (rem_areas - inter) + area[i]  # 2つのエリア（conf最大ボックス+それ以外の各種Box）の総合面積
        IoU = inter/union

        # IoUがoverlapより小さいidxのみを残す
        idx = idx[IoU.le(overlap)]  # leはLess than or Equal toの処理をする演算です
        # IoUがoverlapより大きいidxは、最初に選んでkeepに格納したidxと同じ物体に対してBBoxを囲んでいるため消去

    # whileのループが抜けたら終了

    return keep, count

```

# Detectクラスを実装する

推論時に実行する最後のクラス。(batch_num,21,200,5)サイズの結果を出力する<br>
第一次元：バッチ数を指定する  
第二次元：クラスインデックス  
第三次元：信頼度上位200種のBBoxの何番目か  
第四次元：BBoxの情報(conf,xmin,ymin,width,height)

Detectクラスには,torch.autograph.Functionを継承させる。  
役割
計算グラフの構築 
forward:順伝播  
backward: 逆伝播  
を定義する


```python
#https://qiita.com/windfall/items/073cbb4ffdfab356e495
##自作の関数を定義する際に使用する
class Detect(Function):
    """
    SSDの推論時にconfとlocの出力から、被りを除去したBBoxを出力する

    Param
    =====================
    conf_thresh:float
    conf(BBoxの確信度)の閾値
    top_k:int
    confが高い順にいくつのBBoxを扱うか決定する
    nms_thresh:float
    IOUの閾値を設定する
    """
    def __init__(self, conf_thresh=0.01, top_k=200, nms_thresh=0.45):
        self.softmax = nn.Softmax(dim=-1)  # confをソフトマックス関数で正規化するために用意
        self.conf_thresh = conf_thresh  # confがconf_thresh=0.01より高いDBoxのみを扱う
        self.top_k = top_k  # nm_supression(上で定義した関数)でconfの高いtop_k個を計算に使用する, top_k = 200
        self.nms_thresh = nms_thresh  # nm_supressionでIOUがnms_thresh=0.45より大きいと、同一物体へのBBoxとみなす

    def forward(self, loc_data, conf_data, dbox_list):
        """
        順伝搬の計算を実行する。

        Parameters
        ----------
        loc_data:  [batch_num,8732,4]
            オフセット情報。locモジュールの出力
        conf_data: [batch_num, 8732,num_classes]
            検出の確信度。confモジュールの出力
        dbox_list: [8732,4]
            DBoxの情報。DBox.make_dbox_listの出力

        Returns
        -------
        output : torch.Size([batch_num, 21, 200, 5])
            （batch_num、クラス、confのtop200、BBoxの情報）
        """
　　　　##############TODO:(ここから下の解説付け)##################
        # 各サイズを取得
        num_batch = loc_data.size(0)  # ミニバッチのサイズ
        num_dbox = loc_data.size(1)  # DBoxの数 = 8732
        num_classes = conf_data.size(2)  # クラス数 = 21

        # confはソフトマックスを適用して正規化する
        conf_data = self.softmax(conf_data)

        # 出力の型を作成する。テンソルサイズは[minibatch数, 21, 200, 5]
        output = torch.zeros(num_batch, num_classes, self.top_k, 5)

        # cof_dataを[batch_num,8732,num_classes]から[batch_num, num_classes,8732]に順番変更
        conf_preds = conf_data.transpose(2, 1)

        # ミニバッチごとのループ
        for i in range(num_batch):

            # 1. locとDBoxから修正したBBox [xmin, ymin, xmax, ymax] を求める
            decoded_boxes = decode(loc_data[i], dbox_list)

            # confのコピーを作成
            conf_scores = conf_preds[i].clone()

            # 画像クラスごとのループ（背景クラスのindexである0は計算せず、index=1から）
            for cl in range(1, num_classes):

                # 2.confの閾値を超えたBBoxを取り出す
                # confの閾値を超えているかのマスクを作成し、
                # 閾値を超えたconfのインデックスをc_maskとして取得
                c_mask = conf_scores[cl].gt(self.conf_thresh)
                # gtはGreater thanのこと。gtにより閾値を超えたものが1に、以下が0になる
                # conf_scores:torch.Size([21, 8732])
                # c_mask:torch.Size([8732])

                # scoresはtorch.Size([閾値を超えたBBox数])
                scores = conf_scores[cl][c_mask]

                # 閾値を超えたconfがない場合、つまりscores=[]のときは、何もしない
                if scores.nelement() == 0:  # nelementで要素数の合計を求める
                    continue

                # c_maskを、decoded_boxesに適用できるようにサイズを変更します
                l_mask = c_mask.unsqueeze(1).expand_as(decoded_boxes)
                # l_mask:torch.Size([8732, 4])

                # l_maskをdecoded_boxesに適応します
                boxes = decoded_boxes[l_mask].view(-1, 4)
                # decoded_boxes[l_mask]で1次元になってしまうので、
                # viewで（閾値を超えたBBox数, 4）サイズに変形しなおす

                # 3. Non-Maximum Suppressionを実施し、被っているBBoxを取り除く
                ids, count = nm_suppression(
                    boxes, scores, self.nms_thresh, self.top_k)
                # ids：confの降順にNon-Maximum Suppressionを通過したindexが格納
                # count：Non-Maximum Suppressionを通過したBBoxの数

                # outputにNon-Maximum Suppressionを抜けた結果を格納
                output[i, cl, :count] = torch.cat((scores[ids[:count]].unsqueeze(1),
                                                   boxes[ids[:count]]), 1)

        return output  # torch.Size([1, 21, 200, 5])
```
